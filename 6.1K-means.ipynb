```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets._samples_generator import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import calinski_harabasz_score
```

```python
# 1. 创建数据集
# X为样本特征y为样本簇类别，共1000个样本，每个样本有4个特征，共4个簇
# 簇中心在[-1, -1], [0, 0], [1, 1], [2, 2]，簇方差分别为[0.4, 0.2, 0.2, 0.2]
X, y = make_blobs(n_samples=1000, n_features=2, centers=[[-1, -1], [0, 0], [1, 1], [2, 2]],
                  cluster_std=[0.4, 0.2, 0.2, 0.2], random_state=9)

# 数据集可视化
plt.scatter(X[:, 0], X[:, 1], marker='o')
plt.show()
```

```python
# 2. 使用K-means进行聚类，并使用CH方法评估
y_pred = KMeans(n_clusters=2, random_state=9).fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.show()
# 用CH方法评估聚类得分
print(calinski_harabasz_score(X=X, labels=y_pred))
```

```python
# 2. 使用K-means进行聚类，并使用CH方法评估
y_pred = KMeans(n_clusters=4, random_state=9).fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.show()
# 用CH方法评估聚类得分
print(calinski_harabasz_score(X=X, labels=y_pred))
```

```python
from sklearn.datasets import make_circles

# 生成环状数据
# n_samples: 数据集中的总样本数。
# factor: 控制内外圆环的尺寸比例，取值范围在[0, 1)之间。
# noise: 添加到数据中的高斯噪声的标准差。
X, y = make_circles(n_samples=300, factor=0.5, noise=0.05)

# 绘制生成的数据
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], c='red', label='Class 0')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], c='blue', label='Class 1')

plt.legend()
plt.show()
```

```python
# 2. 使用K-means进行聚类，并使用CH方法评估
y_pred = KMeans(n_clusters=2, random_state=9).fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.show()
# 用CH方法评估聚类得分
print(calinski_harabasz_score(X=X, labels=y_pred))
```
